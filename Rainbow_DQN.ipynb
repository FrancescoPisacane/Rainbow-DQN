## Understanding PufferLib and Ocean Environments

### What is [PufferLib](https://puffer.ai/ocean.html)?

**PufferLib** is a high-performance reinforcement learning framework designed to bridge the gap between research and production RL. It was created to address common pain points in RL development:

- **Speed:** PufferLib environments are highly optimized, often running 10-100x faster than traditional implementations
- **Scalability:** Built-in support for massive parallelization across thousands of environments
- **Simplicity:** Clean, minimal API that follows Gymnasium standards
- **GPU-Native:** Environments can run directly on GPU, eliminating CPU-GPU transfer bottlenecks


### PufferLib Ocean: Educational RL Environments

**Ocean** is PufferLib's collection of lightweight, educational environments. The name reflects its purpose: a "sea" of diverse environments for learning and experimentation. Ocean environments are specifically designed for:

1. **Fast Debugging:** Quickly verify your algorithm works before scaling up
2. **Rapid Prototyping:** Test new ideas without waiting hours for results
3. **Educational Clarity:** Simpler codebases that are easier to understand and modify
4. **Vectorization by Default:** Learn modern RL practices from the start

### Available Ocean Environments

Ocean includes a variety of environments with different characteristics (more can be found in the [docs](https://github.com/PufferAI/PufferLib/tree/3.0/pufferlib/ocean)):

- **Classic Control:** `cartpole`
  - Simple physics simulations
  - Low-dimensional observations
  - Great for debugging and initial testing

- **Grid Worlds:** `squared`, `minigrid_variants`
  - Discrete state/action spaces
  - Fast iteration times
  - Perfect for verifying algorithm correctness

- **Atari-Style:** `breakout`, `pong`, `pacman`, `enduro`
  - More complex visual observations
  - Longer training times
  - Closer to real-world RL challenges

- **Puzzle Games:** `tetris`
  - Strategic planning required

  - Advanced challenge tasks

## 1.2 Environment Factories

````python
class TimeLimitVec:
    """
    Generic vector env time-limit wrapper for PufferLib/Gymnasium-like envs.
    - Marks truncations True when per-env step count reaches max_episode_steps.
    - Preserves existing terminals/truncations from the underlying env.
    - Resets per-env counters on reset or when an env ends.
    """
    def __init__(self, env: Any, max_episode_steps: int, info_key: str = "time_limit"):
        assert max_episode_steps and max_episode_steps > 0
        self.env = env
        self.max_episode_steps = int(max_episode_steps)
        self.info_key = info_key

        # Mirror common attributes for compatibility
        self.single_observation_space = getattr(env, "single_observation_space", None)
        self.single_action_space = getattr(env, "single_action_space", None)
        self.observation_space = getattr(env, "observation_space", None)
        self.action_space = getattr(env, "action_space", None)
        self.num_agents = getattr(env, "num_agents", 1)

        self._steps = np.zeros(self.num_agents, dtype=np.int64)

    def reset(self, seed: Optional[int] = None): 
        if seed is None:
            seed = random.randint(1, 2**32 - 1)
        obs, infos = self.env.reset(seed)
        self._steps[...] = 0
        return obs, infos

    def step(self, actions):
        obs, rewards, terminals, truncations, infos = self.env.step(actions) 
        t = np.asarray(terminals, dtype=bool)
        tr = np.asarray(truncations, dtype=bool)
        if t.ndim == 0:
            t = t.reshape(1)
        if tr.ndim == 0:
            tr = tr.reshape(1)

        active = ~(t | tr)
        # Increment only for envs still active before this step's end flags
        self._steps[active] += 1

        # Apply time limit where not already ended this step
        timeouts = (self._steps >= self.max_episode_steps) & active
        if np.any(timeouts):
            tr = np.logical_or(tr, timeouts)
            self._steps[timeouts] = 0
            if infos is None:
                infos = []
            if not isinstance(infos, list):
                infos = [infos]
            infos.append({self.info_key: {"timeouts": timeouts.copy()}})

        # Also reset counters for any envs that ended naturally
        ended = t | tr
        if np.any(ended):
            self._steps[ended] = 0

        return obs, rewards, t, tr, infos

    def render(self, *args, **kwargs):
        return getattr(self.env, "render")(*args, **kwargs)

    def close(self):
        return getattr(self.env, "close")()
def make_env(name: str, seed: int = 0): 
    env_map = {
        'squared': squared,
        'pong': pong,
        'pacman': pacman,
        'enduro': enduro,
        'tetris': tetris,
        'breakout': breakout,
        'cartpole': cartpole
    }
    if name not in env_map:
        raise ValueError(f'Unknown environment {name}')

    def thunk():
        # Get the module from the map
        env_module = env_map[name]

        # For example, in the 'squared' module, there is a 'Squared' class.
        env_class_name = name.capitalize()
        env_class = getattr(env_module, env_class_name)

        raw_env = env_class(num_envs=1, render_mode=None, seed=seed)  # Instantiate the class
        env = TimeLimitVec(raw_env, max_episode_steps=1000) 
        env.reset(seed=seed)
        return env
    return thunk

debug_env_name = 'squared'
intermediate_env_name = 'cartpole'  # Change to 'pacman', 'enduro', or 'tetris' if desired
challenge_env_name = 'breakout'

test_env = make_env(debug_env_name)()
obs, info = test_env.reset()

print('Debug env observation shape:', np.array(obs).shape)
print('Action space:', test_env.action_space)
print('Observation space:', test_env.observation_space)
````

## 1.4 Replay Buffer


````python
# Replay buffer implementation, a data structure that stores state-action-reward transitions for training the DQN model
class ReplayBuffer:
    def __init__(self, capacity: int, obs_shape, device: torch.device):
        self.capacity = capacity
        self.device = device
        self.pos = 0
        self.full = False

        # Pre-allocate memory 
        self.obs_buf = np.zeros((capacity, *obs_shape), dtype=np.float32)
        self.next_obs_buf = np.zeros((capacity, *obs_shape), dtype=np.float32)
        self.action_buf = np.zeros((capacity,), dtype=np.int64)
        self.reward_buf = np.zeros((capacity,), dtype=np.float32)
        self.done_buf = np.zeros((capacity,), dtype=np.float32)



    def push(self, obs, action, reward, next_obs, done): 
        """Add a new transition to the buffer."""
        self.obs_buf[self.pos] = obs
        self.next_obs_buf[self.pos] = next_obs
        self.action_buf[self.pos] = action
        self.reward_buf[self.pos] = reward
        self.done_buf[self.pos] = done
        self.pos += 1
        if self.pos >= self.capacity:
            self.full = True
            self.pos = 0


    def __len__(self):
        return self.capacity if self.full else self.pos

    def sample(self, batch_size: int):
        """Randomly sample a batch of transitions from the buffer."""
        
    
        current_size = self.__len__()
        
        # Generate random indices 
        indices = np.random.choice(current_size, size=batch_size, replace=True)

        # Retrieve the data at the sampled indices 
        obs = torch.as_tensor(self.obs_buf[indices], dtype=torch.float32, device=self.device)
        next_obs = torch.as_tensor(self.next_obs_buf[indices], dtype=torch.float32, device=self.device)
        actions = torch.as_tensor(self.action_buf[indices], dtype=torch.int64, device=self.device).unsqueeze(1)
        rewards = torch.as_tensor(self.reward_buf[indices], dtype=torch.float32, device=self.device).unsqueeze(1)
        dones = torch.as_tensor(self.done_buf[indices], dtype=torch.float32, device=self.device).unsqueeze(1) 
        
        return obs, actions, rewards, next_obs, dones

# Test ReplayBuffer
print("Testing ReplayBuffer...")
test_buffer = ReplayBuffer(100, (4,), DEVICE) 
test_buffer.push(np.array([1,2,3,4]), 0, 1.0, np.array([2,3,4,5]), False)
assert len(test_buffer) == 1, "Buffer should have 1 element" 
obs, actions, rewards, next_obs, dones = test_buffer.sample(1)
assert obs.shape == (1, 4), f"Expected shape (1, 4), got {obs.shape}" 
print("✓ ReplayBuffer tests passed!")
````

## 1.4 - Extension B - Prioritized Experience Replay

````python
# SumTree data structure used by Prioritized Replay
class SumTree:
    def __init__(self, capacity: int):
        """
        capacity: number of leaf nodes (maximum number of stored transitions)
        The tree is stored in a flat array of size 2 * capacity - 1
        Leaves are in positions [capacity - 1 : 2*capacity - 1)
        """
        self.capacity = capacity
        self.tree = np.zeros(2 * capacity - 1, dtype=np.float32)
        self.data_offset = capacity - 1 

    def _propagate(self, idx: int, change: float):
        """Propagate the change up the tree (update parent sums)."""
        parent = (idx - 1) // 2 
        while True:
            self.tree[parent] += change
            if parent == 0:
                break
            parent = (parent - 1) // 2

    def update(self, leaf_idx: int, new_priority: float):
        """
        Update the priority at the given leaf index (absolute index in tree array).
        leaf_idx must be in range [data_offset, data_offset + capacity)
        """
        change = new_priority - self.tree[leaf_idx]
        self.tree[leaf_idx] = new_priority
        # propagate the change upwards
        if change != 0.0:
            self._propagate(leaf_idx, change)

    def add(self, data_idx: int, priority: float):
        """
        Add or set priority for data index (0..capacity-1).
        This writes to the leaf corresponding to data_idx.
        """
        leaf_idx = data_idx + self.data_offset
        self.update(leaf_idx, priority)

    def get(self, s: float):
        """
        Retrieve the leaf for cumulative sum value s.
        Returns (leaf_index, priority, data_index)
        """
        idx = 0
        while True:
            left = 2 * idx + 1
            right = left + 1
            # if we are at leaf node
            if left >= len(self.tree):
                leaf_idx = idx
                break
            if s <= self.tree[left]:
                idx = left
            else:
                s -= self.tree[left]
                idx = right
        priority = self.tree[leaf_idx]
        data_index = leaf_idx - self.data_offset
        return leaf_idx, priority, data_index

    def total(self) -> float:
        """Return total sum of priorities (value at root)."""
        return float(self.tree[0])

    def min_priority(self, current_size: int) -> float:
        """Return the minimum priority among existing elements (first current_size leaves)."""
        if current_size == 0:
            return 0.0
        leaf_start = self.data_offset
        leaf_end = self.data_offset + current_size
        # Avoid zeros beyond current_size
        return float(np.min(self.tree[leaf_start:leaf_end]))

    def get_leaf_priorities(self, current_size: int) -> np.ndarray:
        """Return array of leaf priorities for existing elements."""
        leaf_start = self.data_offset
        leaf_end = self.data_offset + current_size
        return self.tree[leaf_start:leaf_end].copy()


class PrioritizedReplayBuffer:
    def __init__(self, capacity: int, obs_shape, device: torch.device, alpha: float, eps: float = 1e-6): 
        """
        capacity: max number of transitions
        obs_shape: shape tuple for observations
        device: torch device where tensors should be returned
        alpha: controls how much prioritization is used (0 = uniform, 1 = full prioritized)
        eps: small constant to ensure non-zero priority
        """
        self.capacity = int(capacity)
        self.device = device
        self.pos = 0
        self.full = False
        self.alpha = float(alpha)
        self.eps = float(eps)

        # Pre-allocated memory (same as ReplayBuffer)
        self.obs_buf = np.zeros((self.capacity, *obs_shape), dtype=np.float32)
        self.next_obs_buf = np.zeros((self.capacity, *obs_shape), dtype=np.float32)
        self.action_buf = np.zeros((self.capacity,), dtype=np.int64)
        self.reward_buf = np.zeros((self.capacity,), dtype=np.float32)
        self.done_buf = np.zeros((self.capacity,), dtype=np.float32)

        # SumTree for priorities 
        self.sum_tree = SumTree(self.capacity)

        # Track maximum priority to assign for new transitions
        self.max_priority = 1.0

    def push(self, obs, action, reward, next_obs, done):
        """
        Add a new transition to the buffer and assign it the current max priority
        """
        # store transition in ring buffer
        self.obs_buf[self.pos] = obs
        self.next_obs_buf[self.pos] = next_obs
        self.action_buf[self.pos] = action
        self.reward_buf[self.pos] = reward
        self.done_buf[self.pos] = done

        # set priority for this position (use max_priority so new samples are likely sampled)
        priority = (self.max_priority + self.eps) ** self.alpha 
        self.sum_tree.add(self.pos, priority) 

        # advance pointer
        self.pos += 1
        if self.pos >= self.capacity:
            self.full = True
            self.pos = 0

    def __len__(self) -> int:
        return self.capacity if self.full else self.pos

    def sample(self, batch_size: int, beta: float):
        """
        Sample a batch of transitions according to priority.
        Returns (obs, actions, rewards, next_obs, dones, indices, weights)
        - indices: list/array of data indices (0..capacity-1) that can be used in update_priorities
        - weights: torch tensor of shape (batch_size, 1) for importance-sampling correction
        """
        current_size = len(self)
        if current_size == 0:
            raise ValueError("Cannot sample from an empty buffer")

        # beta: hyperparameter controlling compensation (in [0,1])
        beta = float(beta)

        # Total priority
        total_p = self.sum_tree.total()
        if total_p <= 0.0:
            # fall back to uniform sampling if priorities are zero (it should not happen)
            print("Total priority <= 0, fallback to uniform sampling")
            indices = np.random.choice(current_size, size=batch_size, replace=True)
            probs = np.ones(batch_size) / current_size
        else:
            s_vals = np.random.uniform(0.0, total_p, size=batch_size) # Generate 'batch_size' uniform random values in the range [0, total_p]. These are the 'points' on the cumulative priority line 
            indices = []
            priorities = np.zeros(batch_size, dtype=np.float32)
            for i, s in enumerate(s_vals): # For each sampled value, find the corresponding transition in the SumTree
                _, p, data_index = self.sum_tree.get(s) # Use cumulative sum sampling to find the index of the corresponding transition

                indices.append(int(data_index))
                priorities[i] = p

            probs = priorities / float(total_p) 
            probs = np.maximum(probs, 1e-12) 

        # Compute importance-sampling weights (normalized using batch max weight) 
        N = float(current_size)
        weights = (N * probs) ** (-beta)
        # Normalize by max weight in the current batch
        weights = weights / weights.max()
        weights = weights.astype(np.float32)
        
        indices_arr = np.array(indices, dtype=np.int64) 
       
        obs = torch.as_tensor(self.obs_buf[indices_arr], dtype=torch.float32, device=self.device)
        next_obs = torch.as_tensor(self.next_obs_buf[indices_arr], dtype=torch.float32, device=self.device)
        actions = torch.as_tensor(self.action_buf[indices_arr], dtype=torch.int64, device=self.device).unsqueeze(1)
        rewards = torch.as_tensor(self.reward_buf[indices_arr], dtype=torch.float32, device=self.device).unsqueeze(1)
        dones = torch.as_tensor(self.done_buf[indices_arr], dtype=torch.float32, device=self.device).unsqueeze(1)

        weights_t = torch.as_tensor(weights, dtype=torch.float32, device=self.device).unsqueeze(1)

        return obs, actions, rewards, next_obs, dones, indices_arr, weights_t

    def update_priorities(self, indices: np.ndarray, td_errors: np.ndarray): # Update the priorities of the sampled transitions based on the TD errors calculated during training.
        """
        Update priorities for given indices.
        indices: array-like of data indices (0..capacity-1)
        td_errors: array-like of TD errors (same shape as indices)
        """
        indices = np.asarray(indices, dtype=np.int64)
        td_errors = np.asarray(td_errors, dtype=np.float32)
        assert indices.shape[0] == td_errors.shape[0], "Indices and td_errors must have same length"

        for idx, td in zip(indices, td_errors):
            # priority = (|td| + eps) ** alpha
            p = (abs(float(td)) + self.eps) ** self.alpha
            # update tree leaf
            leaf_idx = int(idx) + self.sum_tree.data_offset
            self.sum_tree.update(leaf_idx, p)
            # maintain max_priority for newly added transitions
            if p > self.max_priority:
                self.max_priority = float(p)

    def get_all_priorities(self) -> np.ndarray:
        """Return leaf priorities array (length == capacity)."""
        return self.sum_tree.tree[self.sum_tree.data_offset:self.sum_tree.data_offset + self.capacity].copy()
````

## 1.5 Epsilon Schedule
Linear decay from start to end over `epsilon_decay_steps`.

````python
#Function to calculate the epsilon value based on the number of steps taken, used in the epsilon-greedy strategy to balance exploration and exploitation during the DQN model training.
def epsilon_by_step(step: int, cfg: Config) -> float:
    if step < 0:
        return cfg.epsilon_start 
    elif step >= cfg.epsilon_decay_steps:
        return cfg.epsilon_end
    elif 0 <= step < cfg.epsilon_decay_steps:
        epsilon = cfg.epsilon_start - (cfg.epsilon_start - cfg.epsilon_end) * (step / cfg.epsilon_decay_steps)
        return epsilon

    return None

print('Epsilon(0)=', epsilon_by_step(0, cfg), 'Epsilon(mid)=', epsilon_by_step(cfg.epsilon_decay_steps//2, cfg), 'Epsilon(end)=', epsilon_by_step(cfg.epsilon_decay_steps*2, cfg))
````

## 1.6 Q-Network Architectures

````python
# noisy network 

import math

class NoisyLinear(nn.Module):
    def __init__(self, in_features, out_features, sigma_init=0.017):
        super(NoisyLinear, self).__init__()
        self.in_features = in_features
        self.out_features = out_features
        
        # Standard parameters
        self.weight_mu = nn.Parameter(torch.empty(out_features, in_features))
        self.bias_mu = nn.Parameter(torch.empty(out_features))
        
        # Noise parameters
        self.weight_sigma = nn.Parameter(torch.full((out_features, in_features), sigma_init))
        self.bias_sigma = nn.Parameter(torch.full((out_features,), sigma_init))
        
        # Buffers for noise
        self.register_buffer("weight_epsilon", torch.zeros(out_features, in_features))
        self.register_buffer("bias_epsilon", torch.zeros(out_features))
        
        self.reset_parameters()
        self.reset_noise()

    def reset_parameters(self):
        mu_range = 1 / math.sqrt(self.in_features)
        self.weight_mu.data.uniform_(-mu_range, mu_range)
        self.bias_mu.data.uniform_(-mu_range, mu_range)

    def reset_noise(self):
        self.weight_epsilon.normal_()
        self.bias_epsilon.normal_()

    def forward(self, x):
        if self.training:
            w = self.weight_mu + self.weight_sigma * self.weight_epsilon
            b = self.bias_mu + self.bias_sigma * self.bias_epsilon
        else:
            w = self.weight_mu
            b = self.bias_mu
        return torch.nn.functional.linear(x, w, b)
````

````python
# Implementation of the neural network to approximate the Q function, which estimates the expected value of future rewards for every possible action in a given state.

class MLPQ(nn.Module):
    def __init__(self, obs_dim: int, action_dim: int, dueling=False):
        FinalLayer = NoisyLinear if cfg.use_noisy else nn.Linear
        super().__init__()
        self.hidden_size = cfg.hidden_size # tunable
        self.dueling = dueling


        # Standard DQN architecture
        if not dueling:
            self.net = nn.Sequential(
                nn.Linear(obs_dim, self.hidden_size),
                nn.ReLU(),
                nn.Linear(self.hidden_size, self.hidden_size),
                nn.ReLU(),
                FinalLayer(self.hidden_size, action_dim)
            )
        else: 
            # Dueling DQN architecture 
            self.feature_layer = nn.Sequential( 
                nn.Linear(obs_dim, self.hidden_size),
                nn.ReLU()
            )
            self.value_stream = nn.Sequential(
                nn.Linear(self.hidden_size, self.hidden_size),
                nn.ReLU(),
                FinalLayer(self.hidden_size, 1)
            )
            self.advantage_stream = nn.Sequential(
                nn.Linear(self.hidden_size, self.hidden_size),
                nn.ReLU(),
                FinalLayer(self.hidden_size, action_dim)
            )


    def forward(self, x): 
        if not self.dueling:
            return self.net(x)
        else:
            features = self.feature_layer(x)
            values = self.value_stream(features)
            advantages = self.advantage_stream(features)
            q_values = values + (advantages - advantages.mean(dim=1, keepdim=True)) 
            return q_values 
        
    def reset_noise_all(self):
        """Reset the noise in all NoisyLinear layers of the network."""
        if not cfg.use_noisy:
            return

        if not self.dueling:
            if hasattr(self.net[-1], 'reset_noise'):
                self.net[-1].reset_noise()
        else:
            if hasattr(self.value_stream[-1], 'reset_noise'):
                self.value_stream[-1].reset_noise()
            if hasattr(self.advantage_stream[-1], 'reset_noise'):
                self.advantage_stream[-1].reset_noise()



def build_q_network(env, cfg: Config):
    obs_space = env.single_observation_space
    print('Obs space:', obs_space.shape[0])
    act_space = env.single_action_space
    action_dim = act_space.n
    assert isinstance(act_space, gym.spaces.Discrete), 'DQN requires discrete actions'
    if len(obs_space.shape) == 1:
        return MLPQ(obs_space.shape[0], action_dim, dueling=cfg.use_dueling)


# Quick test on debug env
net_test_env = test_env
test_net = build_q_network(net_test_env, cfg).to(DEVICE)
print(test_net)
````

## 1.7 Action Selection
Standard epsilon-greedy policy. (Later you might incorporate noisy networks or exploration bonuses.)

````python
# Implementation of the function to select an action using an epsilon-greedy strategy.

@torch.no_grad()
def select_action(q_net, obs, epsilon: float, env): 
    if random.random() < epsilon: 
        return env.action_space.sample() 
    if isinstance(obs, torch.Tensor): 
        obs_t = obs.to(DEVICE, dtype=torch.float32, non_blocking=True) 
    else:
        obs_t = torch.as_tensor(obs, dtype=torch.float32, device=DEVICE)
    q_vals = q_net(obs_t)
    return int(q_vals.argmax(dim=1).item()) 
    
# Test call
dummy_action = select_action(test_net, obs, 1.0, test_env)
print('Random/greedy test action:', dummy_action)
````

## 1.8 Training Step
Compute standard DQN loss or Double DQN target if `cfg.use_double_dqn=True`.

Target for vanilla DQN:
$$ y = r + (1-d) \gamma \max_{a'} Q_{target}(s', a') $$

("d" is a flag for the end of the episode)

Double DQN variant:
$$ a^* = \arg\max_{a'} Q_{online}(s', a') \quad; \quad y = r + (1-d) \gamma Q_{target}(s', a^*) $$

````python


def train_step(q_net, target_net, optimizer, replay: ReplayBuffer, cfg: Config):
    if len(replay) < cfg.min_buffer_size: 
        return None

    if cfg.use_prioritized_experience_replay:
        try:
            obs, actions, rewards, next_obs, dones, indices, weights = replay.sample(cfg.batch_size, cfg.beta)
        except TypeError:
            print("use_prioritized_experience_replay is true, but replay is not a PrioritizedReplayBuffer")
    else:
        try:
            obs, actions, rewards, next_obs, dones = replay.sample(cfg.batch_size)
        except TypeError:
            print("use_prioritized_experience_replay is false, but replay is not a ReplayBuffer")
        
    
    q_values = q_net(obs) # Forward pass to get the estimated Q values for the current state.
    q_sa = q_values.gather(1, actions) 
    
    with torch.no_grad():
        if cfg.use_double_dqn:
            # Double DQN
            next_q_values_online = q_net(next_obs) 
            next_actions = next_q_values_online.argmax(dim=1, keepdim=True) 
            next_q_values_target = target_net(next_obs) 
            max_next_q_values = next_q_values_target.gather(1, next_actions) 

        else:
            # Vanilla DQN
            next_q_values = target_net(next_obs) 
            max_next_q_values, _ = next_q_values.max(dim=1, keepdim=True)

        if cfg.use_n_step: target = rewards + (1 - dones) * (cfg.gamma ** cfg.n_step) * max_next_q_values
        else: target = rewards + (1 - dones) * cfg.gamma * max_next_q_values

        if cfg.reward_clip is not None:
            target = torch.clamp(target, -cfg.reward_clip, cfg.reward_clip) 

    loss_fn = nn.SmoothL1Loss(reduction='none')  # compute element-wise Huber losses, mean computed later

    loss = loss_fn(q_sa, target)  # shape [batch_size, 1]
    loss = loss.squeeze(1)       # flatten

    if cfg.use_prioritized_experience_replay:
        weights = weights.squeeze(1)
        loss = (loss * weights)   # apply IS weights

    loss = loss.mean()            # reduce to scalar

    optimizer.zero_grad() 
    loss.backward() 
    nn.utils.clip_grad_norm_(q_net.parameters(), cfg.gradient_clip) 
    optimizer.step()

    # If PER was used, compute TD-errors and update priorities
    if cfg.use_prioritized_experience_replay:
        # TD-errors: absolute difference between target and current Q
        with torch.no_grad():
            td_errors = (target - q_sa).abs().squeeze()  # shape (batch_size,)
            # Convert to cpu numpy for update_priorities (implementation expects numpy)
            td_errors_np = td_errors.cpu().numpy()
            # Update priorities in the replay buffer
            replay.update_priorities(indices, td_errors_np)

    return float(loss.item()) 
````

## 1.9 Reward Processing
Add optional reward clipping to stabilize training on environments with high or varied reward magnitudes.

If `cfg.reward_clip` is set, clip reward to `[-cfg.reward_clip, cfg.reward_clip]`.

````python

def process_reward(r: float, cfg: Config):
    if cfg.reward_clip is not None:
        return max(-cfg.reward_clip, min(cfg.reward_clip, r)) 
    return r

# Quick test
cfg.reward_clip = None  # Set to 1.0 to try clipping later
print('Reward test (no clip):', process_reward(5.0, cfg))
cfg.reward_clip = 1.0
print('Reward test (clip=1.0):', process_reward(5.0, cfg))
cfg.reward_clip = None  # Reset for training; modify if desired
````

## 1.10 Evaluation Utilities
Evaluation runs with greedy policy (`epsilon=0`).


````python
# Function to evaluate the performance of the DQN model in a specific environment, running a certain number of episodes and calculating the average reward obtained

def evaluate(env_fn, q_net, cfg: Config, episodes: int):
    env = env_fn
    returns = []
    for _ in range(episodes):
        obs, _ = env.reset()
        done = False
        ep_ret = 0
        steps = 0
        while not done and steps < cfg.max_episode_len:
            a = select_action(q_net, obs, 0.0, env) 
            obs2, r, terminated, truncated, _ = env.step(a)
            done = terminated or truncated
            ep_ret += r
            obs = obs2
            steps += 1
        returns.append(ep_ret)
    return float(np.mean(returns)) 

def hard_update(target, source): # Copy the weights from the source network to the target network, useful for periodically updating the target network with the weights of the online network
    target.load_state_dict(source.state_dict())

print('Eval utility ready.')
````

````python
# Functions to save and load the parameters of the DQN model, including network weights, optimizer state, and training configuration.

def save_checkpoint(path, q_net, optimizer, cfg: Config, stage_name: str):
    torch.save({
        'model': q_net.state_dict(),
        'optimizer': optimizer.state_dict(),
        'config': dataclasses.asdict(cfg),
        'stage': stage_name,
    }, path)
    print('Saved checkpoint:', path)

def load_checkpoint(path, q_net, optimizer):
    data = torch.load(path, map_location=DEVICE)
    q_net.load_state_dict(data['model'])
    optimizer.load_state_dict(data['optimizer'])
    print('Loaded checkpoint from', path)
````

````python
## Used to save logs in json format
import json
from pathlib import Path

def save_logs_to_file(logs, filename):
    """Save logs dict to filename as JSON. Overwrites if file exists."""
    # Ensure parent dir exists
    p = Path(filename)
    if p.parent != Path("."):
        p.parent.mkdir(parents=True, exist_ok=True)
    with open(p, "w") as f:
        json.dump(logs, f, indent=2)
````

````python
def compute_n_step_return(buffer, gamma):
    R = 0.0
    for i, (_, _, reward, _, done) in enumerate(buffer):
        R += (gamma ** i) * reward
        if done:
            break
    next_state = buffer[-1][3]
    done_flag = buffer[-1][4]
    return R, next_state, done_flag
````

````python
all_stage_logs: Dict[str, Dict[str, List]] = {}

q_net = None
target_net = None
optimizer = None
stage_index = 0

# Easy way to check agent performance over different envs
for stage_env_name in cfg.curriculum:
    print(f"\n=== Stage {stage_index+1}: {stage_env_name} ===")
    env = make_env(stage_env_name, seed=cfg.seed + stage_index)()
    eval_env = make_env(stage_env_name, seed=cfg.seed + 100 + stage_index)()


    q_net = build_q_network(env, cfg).to(DEVICE)
    target_net = build_q_network(env, cfg).to(DEVICE)
    hard_update(target_net, q_net)
    optimizer = optim.Adam(q_net.parameters(), lr=cfg.lr)

    n_step_buffer = None
    if cfg.use_n_step:
        n_step_buffer = deque(maxlen=cfg.n_step) # Initialization buffer n-step

    if cfg.use_prioritized_experience_replay:
        replay = PrioritizedReplayBuffer(cfg.buffer_capacity, env.single_observation_space.shape, DEVICE, alpha = cfg.alpha, eps = 1e-6)
    else:
        replay = ReplayBuffer(cfg.buffer_capacity, env.single_observation_space.shape, DEVICE)

    logs = {
        'episode_rewards': [],
        'losses': [],
        'eps_history': [],
        'eval': [],
    }

    obs, _ = env.reset()
    ep_reward = 0
    ep_steps = 0
    episode_counter = 0
    recent_ep_returns = []
    total_steps = 0
    print('Starting training...')
    while total_steps < cfg.max_steps_per_env:
        if cfg.use_noisy:
            epsilon = 0  
            q_net.reset_noise_all() 
            target_net.reset_noise_all()
        else:
            epsilon = epsilon_by_step(total_steps, cfg)
            
        action = select_action(q_net, obs, epsilon, env) 
        next_obs, reward, terminated, truncated, info = env.step(action)
        reward = process_reward(reward, cfg)
        reward = float(np.asarray(reward).reshape(-1)[0])
        done = bool(np.asarray(terminated).reshape(-1)[0] or np.asarray(truncated).reshape(-1)[0])
        if cfg.use_n_step:
          
            n_step_buffer.append((obs.copy(), int(action), reward, next_obs.copy(), done))
            if len(n_step_buffer) == cfg.n_step:
                R_n, next_state_n, done_n = compute_n_step_return(n_step_buffer, cfg.gamma)
                first_state, first_action = n_step_buffer[0][0], n_step_buffer[0][1]
                replay.push(first_state, first_action, R_n, next_state_n, done_n)
                n_step_buffer.popleft() # Remove the first element
        else:
            # Logic 1-step standard
            replay.push(obs.copy(), int(action), reward, next_obs.copy(), done)

        obs = next_obs.copy()
        ep_reward += reward
        ep_steps += 1
        total_steps += 1

        # Training
        if total_steps % cfg.train_freq == 0:
            loss = train_step(q_net, target_net, optimizer, replay, cfg)
            if loss is not None:
                logs['losses'].append(loss)

        # Target update
        if total_steps % cfg.target_update_interval == 0:
            hard_update(target_net, q_net)

        logs['eps_history'].append(epsilon)

        if done or ep_steps >= cfg.max_episode_len:
            logs['episode_rewards'].append(ep_reward)
            recent_ep_returns.append(ep_reward)
            episode_counter += 1
            if episode_counter % cfg.log_interval_episodes == 0:
                mean_recent = np.mean(recent_ep_returns[-cfg.log_interval_episodes:])
                print(f"[{stage_env_name}] Ep {episode_counter} Steps {total_steps} RecentMean {mean_recent:.2f} Eps {epsilon:.3f}")
            obs, _ = env.reset()
            ep_reward = 0
            ep_steps = 0

            # Periodic evaluation (every 2*log_interval episodes if we have progress)
            if episode_counter % (cfg.log_interval_episodes * 2) == 0:
                eval_ret = evaluate(eval_env, q_net, cfg, cfg.eval_episodes)
                logs['eval'].append((total_steps, eval_ret))
                print(f"Eval @ Ep {episode_counter} and {total_steps} steps => mean return {eval_ret:.2f}")                 

        
    # save the weights if needed
    if cfg.saveWeights:
        CHECKPOINT_PATH = f"{stage_env_name}_network_weights.pt"
        save_checkpoint(CHECKPOINT_PATH, q_net, optimizer, cfg, stage_env_name)

    # Save the logs if needed
    if cfg.SAVE_LOGS_NAME is not None:
        SAVE_LOGS_PATH = f"results/{stage_env_name}_{cfg.SAVE_LOGS_NAME}_logs.json"
        save_logs_to_file(logs, SAVE_LOGS_PATH)
        
    all_stage_logs[stage_env_name] = logs
    stage_index += 1
````

## Multiple training run visualization

The following cell defines the `plot_logs_overlay` function, used to load and overlay the log files (JSON) from multiple reinforcement learning training runs.

It generates a 1x2 subplot showing **Rewards** (a smoothed moving average) and **Evaluation** results (Mean Return vs. Env Steps).

### Note on Reward Plot Normalization

Training runs are fixed by a total number of **steps**, not episodes. This means a "better" agent (which survives longer) will complete *fewer* total episodes within the same step budget compared to a "worse" agent (which dies frequently).

To make a fair comparison, the **Reward** plot's x-axis (episode count) is normalized to a 0.0-1.0 range for each run. This allows us to compare the relative learning progress from the start (0.0) to the end (1.0) of training, regardless of the absolute number of episodes completed in each run.

````python
import json
from pathlib import Path
from typing import List, Optional
import numpy as np
import matplotlib.pyplot as plt

def moving_average(x, w):
    x = np.array(x).flatten()
    if w is None or len(x) < w:
        return x
    return np.convolve(x, np.ones(w)/w, mode='valid')

def plot_rewards_eval_overlay(
    paths: List[str],
    labels: Optional[List[str]] = None,
    smoothing_rewards: int = 100,
    max_episodes: Optional[int] = None,
    save_path: Optional[str] = None,
):
    """
    Overlay multiple training runs: Rewards (smoothed) and Evaluation Returns
    in two side-by-side subplots.
    
    NOTE: Rewards X-axis is Normalized Episodes (0-1).
    Evaluation X-axis is total environment Steps.
    """
    if labels is None:
        labels = [Path(p).stem for p in paths]
    if len(labels) != len(paths):
        raise ValueError("labels must be the same length as paths or None")

    # Load all runs
    runs = []
    for p in paths:
        p = Path(p)
        if not p.exists():
            print(f"Warning: file not found: {p}")
            runs.append(None)
            continue
        with open(p, "r") as f:
            data = json.load(f)

        rewards, evals = [], []
        if isinstance(data, dict) and "episode_rewards" in data:
            rewards = list(data.get("episode_rewards", []))
            evals = list(data.get("eval", []))
        elif isinstance(data, dict):
            # Aggregate data if logs are saved by stage (e.g., curriculum)
            for k in sorted(data.keys()):
                v = data[k]
                if not isinstance(v, dict):
                    continue
                rewards.extend(v.get("episode_rewards", []))
                evals.extend(v.get("eval", []))

        runs.append({"rewards": rewards, "evals": evals})

    # Colors
    prop_cycle = plt.rcParams['axes.prop_cycle']
    colors = prop_cycle.by_key()['color']
    color_map = {lbl: colors[i % len(colors)] for i, lbl in enumerate(labels)}

    # Figure with two side-by-side subplots
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    ax_r, ax_eval = axes[0], axes[1]
    
    # === Subplot for Rewards ===
    ax_r.set_title("Smoothed Episode Rewards (Normalized)")
    ax_r.set_xlabel("Normalized Episodes (0-1)")
    ax_r.set_ylabel("Mean Return (MA)")
    ax_r.grid(True)
    
    # === Subplot for Evaluation Returns ===
    ax_eval.set_title("Evaluation Returns")
    ax_eval.set_xlabel("Env Steps")
    ax_eval.set_ylabel("Mean Return (Greedy)")
    ax_eval.grid(True)


    for run, lbl in zip(runs, labels):
        if run is None:
            continue
        color = color_map[lbl]

        # Rewards (Plot on the first subplot - Normalized X-axis)
        rewards = run["rewards"]
        if max_episodes:
            rewards = rewards[:max_episodes]
        if len(rewards) > 0:
            ma = moving_average(rewards, smoothing_rewards)
            start_x_idx = len(rewards) - len(ma)
            # Normalize the X-axis for episodes (0-1)
            x_norm_r = np.linspace(0, 1, len(rewards))[start_x_idx:]
            ax_r.plot(x_norm_r, ma, label=f"{lbl} MA({smoothing_rewards})", color=color)

        # Evaluation returns (Plot on the second subplot - X-axis with Steps)
        evals = run["evals"]
        if evals:
            try:
                # Extract steps (X-axis) and returns (Y-axis)
                xs = [float(s) for s, _ in evals]
                ys = [float(r) for _, r in evals]
                ax_eval.plot(xs, ys, marker="o", linestyle="-", color=color, label=f"{lbl}", linewidth=1)
            except Exception as e:
                print(f"Error plotting evaluation for {lbl}: {e}")
                continue

    # Legend (one per subplot)
    if len(paths) > 1:
        ax_r.legend(fontsize="small", loc="upper left")
        ax_eval.legend(fontsize="small", loc="upper left")

    plt.tight_layout()
    if save_path:
        out_path = Path(save_path)
        out_dir = out_path.parent
        if out_dir != Path("."):
            out_dir.mkdir(parents=True, exist_ok=True)
        fig.savefig(str(out_path))
        plt.close(fig)
        print(f"Saved reward and eval figure to {out_path}")
    else:
        plt.show()
````

````python
# Example usage: visualize rewards and evaluation returns only

paths = [
    r"results/cartpole_Vanilla_logs.json",
    r"results/cartpole_ALL_logs.json",
    r"results/cartpole_ALL_tent1_logs.json",
    r"results/cartpole_ALL_tent3_logs.json"
]

labels = [
    "Vanilla DQN  ",
    "All extensions  ",
    "ALL + Found Params",
    "All + Good Params  "

] if len(paths) > 1 else None 

output_filename = None

# Call the function to plot
plot_rewards_eval_overlay(
    paths=paths,
    labels=labels,
    smoothing_rewards=100,      # moving average window for rewards
    max_episodes=None,
    save_path=output_filename
)
````
